```python
def tort_eval(gold_hin, pred_hin, rew_list):
    assert(len(gold_hin)==len(pred_hin))
    macro_total = 0
    macro_pred = 0
    micro_total = 0
    for idx, gold_sent in enumerate(gold_hin):
        micro = 0
        micro_pred = 0
        for rew in rew_list:
            rew_count = gold_sent.count(rew)
            if rew_count>0:
                macro_total += rew_count
                micro += rew_count
                rew_pred = pred_hin[idx].count(rew)
                macro_pred += min(rew_pred, rew_count)
                micro_pred += min(rew_pred, rew_count)
        if(micro!=0):
            micro_total += micro_pred/micro
    
    macro_eval = macro_pred/macro_total
    micro_eval = micro_total/len(gold_hin)
    return macro_eval, micro_eval
                
        
```
