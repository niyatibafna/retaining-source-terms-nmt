```python
import re
def tort_eval(gold_hin, pred_hin, rew_list):
    assert(len(gold_hin)==len(pred_hin))
    #find recall
    macro_total = 0
    macro_pred = 0
    micro_total = 0
    
    macro_total_prec = 0
    macro_pred_prec = 0
    micro_total_prec = 0
    
    for idx, gold_sent in enumerate(gold_hin):
        micro = 0
        micro_pred = 0
        
        for rew in rew_list:
            rew_count = gold_sent.count(rew)
            if rew_count>0:
                macro_total += rew_count
                micro += rew_count
                rew_pred = pred_hin[idx].count(rew)
                macro_pred += min(rew_pred, rew_count)
                micro_pred += min(rew_pred, rew_count)
        if(micro!=0):
            micro_total += micro_pred/micro
            
        micro_prec = 0
        micro_pred_prec = 0
        pred_sent = pred_hin[idx]
        eng_words = re.findall(" [a-zA-Z]* ", pred_sent)
        for eng in eng_words:
            macro_total_prec += pred_sent.count(eng)
            micro_prec += pred_sent.count(eng)
            micro_pred_prec += min(gold_sent.count(eng), pred_sent.count(eng)) if eng.strip() in rew_list else 0
            macro_pred_prec += min(gold_sent.count(eng), pred_sent.count(eng)) if eng.strip() in rew_list else 0
            
            
        if micro_prec!=0:
            micro_total_prec += micro_pred_prec/micro_prec
            
            
    
    macro_rec = macro_pred/macro_total
    micro_rec = micro_total/len(gold_hin)
    
    macro_prec = macro_pred_prec/macro_total_prec
    micro_prec = micro_total_prec/len(gold_hin)
    
    micro_f1 = 2 * (micro_prec * micro_rec) / (micro_prec + micro_rec)
    macro_f1 = 2 * (macro_prec * macro_rec) / (macro_prec + macro_rec)
    
    
    print(micro_prec, "\t", micro_rec, "\t", micro_f1)
    print(macro_prec, "\t", macro_rec, "\t", macro_f1)
    return micro_f1, macro_f1
                
        
```

```python
#unseen
rew_list = open("../stats/unseen_rew_list.txt","r").read().strip().split("\n")
gold_hin = open("../test_outputs/tagged/test.unseen.hindi", "r").read().strip().split("\n")
pred_hin_tagged = open("../test_outputs/tagged/test.unseen.out", "r").read().strip().split("\n")
pred_hin_baseline = open("../test_outputs/baseline/test.unseen.out", "r").read().strip().split("\n")
print(tort_eval(gold_hin, pred_hin_tagged, rew_list))
print(tort_eval(gold_hin, pred_hin_baseline, rew_list))
```

```python
#seen
seen_rew_list = open("../stats/seen_rew_list.txt","r").read().split("\n")
seen_rew_list = [x.strip() for x in seen_rew_list]
gold_hin = open("../test_outputs/tagged/test.seen.hindi", "r").read().strip().split("\n")
pred_hin_tagged = open("../test_outputs/tagged/test.seen.out", "r").read().strip().split("\n")
pred_hin_baseline = open("../test_outputs/baseline/test.seen.out", "r").read().strip().split("\n")
print(tort_eval(gold_hin, pred_hin_tagged, seen_rew_list))
print(tort_eval(gold_hin, pred_hin_baseline, seen_rew_list))
```
